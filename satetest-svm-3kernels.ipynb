```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import os
from tqdm import tqdm
```

```python
#data directory path
DATADIR = '../input/satellite-image-classification/data'
CATEGORIES = ['water', 'cloudy', 'green_area', 'desert']
IMG_SIZE=100
```

```python
#preprocessing the data and creation of training dataset
training_data=[]
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR, category)
        class_num=CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img))
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
   
```

```python
create_training_data()
```

```python
print(len(training_data))
```

```python
lenofimage = len(training_data)
```

```python
#we need to convert the images to array form for our SVM model to train on it
X=[]
y=[]

for categories, label in training_data:
    X.append(categories)
    y.append(label)
X= np.array(X).reshape(lenofimage,-1) #since SVM takes 2D input to train
```

```python
X.shape
```

```python
#this is to flatten the array
X = X/255.0
```

```python
#same for y data
y=np.array(y)
```

```python
y.shape
```

```python
#applying train_test_split on our data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
#creation SVM model and fitting our data
from sklearn.svm import SVC
svc = SVC(kernel='linear',gamma='auto')
svc.fit(X_train, y_train)
```

```python
#predicting the X_test
y2 = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy on given data with Linear Kernel",accuracy_score(y_test,y2))
```

```python
from sklearn.metrics import classification_report
print("Accuracy on given data with Linear Kernel \n",classification_report(y_test,y2))
```

```python
#fitting data in SVM model with rbf kernel (radial basis function) 
from sklearn.svm import SVC
svc = SVC(kernel='rbf',gamma=0.5)
svc.fit(X_train, y_train)
```

```python
#predicting with X_test again
y3 = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy on given data with rbf kernel",accuracy_score(y_test,y3))
```

```python
from sklearn.metrics import classification_report
print("Accuracy on given data with rbf Kernel \n",classification_report(y_test,y3))
```

```python
#fitting data in SVM model with poly kernel
from sklearn.svm import SVC
svc = SVC(kernel='poly',degree=4)
svc.fit(X_train, y_train)
```

```python
#predicting with X_test again
y4 = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy on given data with rbf kernel",accuracy_score(y_test,y4))
```

```python
from sklearn.metrics import classification_report
print("Accuracy on given data with poly Kernel \n",classification_report(y_test,y4))
```
